# Burn Severity Assessment

In this project, I mapped the impact of the 2017 wildfire in Arkansas by visualizing the burn severity.

### Computing Platform & Methodology
---
- Google Earth Engine ([Link to GEE Script](https://code.earthengine.google.com/c5e8d4d2d1e51182fff85fa00b19b512?noload=1))
- Delta Normalized Burn Ratio

### Data Sources
---
- [Landsat 8 imagery](https://developers.google.com/earth-engine/datasets/catalog/landsat-8/)
  - Pre-fire: January 1, 2017 – June 30, 2017
  - Post-fire: July 1, 2017 – December 31, 2017

### Process Overview
- Define the area of interest (-In my case, I imported a shapefile of Arkansas boundary-)
- Import the Landsat 8 imagery and filter to date range
- Apply cloud and snow masks to ensure clarity and scale reflectance values
- Mosaic Images: Combine multiple images into single pre- and post-fire composites.
- Compute the Normalized Burn Ratio (NBR) for both pre- and post-fire images.
- Derive the delta NBR (dNBR), which quantifies the change in burn severity.
