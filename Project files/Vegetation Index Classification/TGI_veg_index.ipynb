# Vegetation Index Classification of UAV-Captured Imagery

This is one of my capstone projects for the EdX Massive Open Online Course (MOOC) titled "Drone for Agriculture," organized by Wageningen University & Research (WUR).

---
### Project
In this project, a group of colleagues is evaluating the performance of various fertilizers on a small research field with several quinoa patches, where different products have been applied. To facilitate data access and monitoring throughout the process, aerial imaging of the field is being used.<br>
I have been tasked with processing the Orthomosaic and visualizing the Triangular Greenness Index (TGI) on a map.


### Data Source
- UAV-Captured Imagery<BR>
  - Download details &nbsp;([link](https://mooc.wur.nl/downloads/odm_orthophoto.tif) |&nbsp;&nbsp;Username: moocuser |&nbsp; Password: Dr0n3Fl1ght!)

### Software
- QGIS V3.3.6

### Methodology
- Download the imagery
- Import the Orthomosaic into QGIS
- Calculate the Triangular Greenness Vegetation Index (in QGIS Raster Calculator)
  TGI = G - 0.39R - 0.61B

- Digitize boundary of interest, visualize with a meaningful color scale and create a final layout
